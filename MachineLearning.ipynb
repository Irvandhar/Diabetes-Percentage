TAHAPAN KLASISIFKASI MENGGUNAKAN NAIVE BAYES

```python
#Memasukkan algoritma naive bayes
from sklearn.naive_bayes import GaussianNB

#Memasukkan model gauissian
NBmodel = GaussianNB()
```

```python
#Memasukkan data variabel independen kedalam var X
X = data[['glucose', 'bloodpressure']]
X
```

```python
# Memasukkan variabel dependen ke dalam variabel y
y = data["diabetes"]
y
```

```python
#import Library machine Learning sklearn
from sklearn.model_selection import train_test_split
```

```python
#Melatih data X dan y
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 123)
```

```python
x_test.head(5)
```

```python
x_test.shape
```

```python
nbtrain = NBmodel.fit(x_train, y_train)
nbtrain.class_count_
```

```python
#prediksi
y_pred = nbtrain.predict(x_test)
y_pred
```

```python
Prob = nbtrain.predict_proba(x_test)
Prob.shape
```
